### 1. Handling Missing Values in Airbnb Listings Bangkok

```python
#Import Libraries
import pandas as pd
import numpy as np
```

```python
#Open Data Set
df = pd.read_csv('Airbnb Listings Bangkok.csv')
df.head()
```

```python
#Check Missing Values
df.isna().sum()
```

```python
#Check Dataframe Info
df.info()
```

### 1.1 Filling missing values in column 'name'

```python
#Filling missing value in column name
df['name'].fillna(value = 'noname', inplace=True)
```

```python
df.isna().sum()
```

### 1.2 Filling missing values in column 'host_name'

```python
#Filling missing values in column host_name
df['host_name'].fillna(value = 'noname', inplace=True)
```

```python
df.isna().sum()
```

### 1.3 Filling missing values in column 'last_review'

```python
#Filling missing values in column last_review

df['last_review'].fillna(value = '1945-01-01', inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

### 1.4 Filling missing values in column 'reviews_per_month'

```python
df['reviews_per_month'].fillna(value = df['reviews_per_month'].mean(), inplace=True)
```

```python
df.isna().sum()
```
